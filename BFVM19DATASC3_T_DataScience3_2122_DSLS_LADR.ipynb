###### batch_size = 64
The batch size is the number of samples that are passed from the generator to the network at once. It defines the number of samples to work through before updating the internal model parameters.
too large of a batch size will lead to poor generalization.
I chose 64 insteadd of defualt value(32) because of large training dataset and greater batch sizes result in faster progress in training. ### Convolutional Neural Network
Convolutional neural network is better than a traditional neural network because CNN has features parameter sharing and dimensionality reduction.The number of parameters is decreased thus the computations also reduces.
filters or channels in convolutional NN are automatically able to detect patterns in images.

paterns can be multiple shapes, edges, textures, objects, colors,etc. some filter detects shape.

The deeper the network goes, filters become more expert. Filters are able to detect specific objects like eyes, ears, hair or fur and scales in deeper layers.

######  filter
filter extracts features from the matrix of data; the number of channels should be low in the beginning because it detects low-level features that are combined to collect local information as much as possible and then by increasing the number of channels distinguishes between classes. it is better to define number of filters is power of two because there are characteristics of hardware that work more efficiently when memory chunks stored in powers of two are passed.

###### kernel_size = (3,3)
kernel_size specify the height and width of convolution window. as data are very similar to each others(there are some wild animals that are same as cat and dogs), I chose kernel_size = (3,3) to ensure most information 
###### activation = "relu"
at first I tried the model bu sofplus activation function and the result was noisy; then I tried by "relu" and the result was good;

###### number of parameters:
param_number = output_channel_number * (input_channel_number * kernel_height * kernel_width + 1)
it is written in code for every layers.


#### MaxPooling2D
the number of parameters for all MaxPooling2D layers is all 0. The reason is that this layer doesnt learn anything. it reduces the complexity of the model and to extract local features by finding the maximum values for each 2 x 2(default) pool.

###### pool_size:
window size over which to take the maximum from previous step as I wanted to take the as much data and miss less data.
###### strides
Stride is the number of pixels movement over the input matrix. I set it to (2,2) and not higher to pass as much as possible data to next layer.
###### padding
1.  valid
the input image is not padded. This means that the filter window always stays inside the input image. there can be a loss of information. How many elements are ignored depends on the size of the kernel and the stride.
2. same
when we think the image borders may be important we set padding on "same" and filters are applied to all input nodes.

```python
animal_trad = keras.models.Sequential()
animal_trad.add(keras.layers.InputLayer(input_shape=(126, 126, 3)))
animal_trad.add(keras.layers.Flatten())
animal_trad.add(keras.layers.Dense(units = 64, activation='relu'))
# 64*(126*126*3+1)
animal_trad.add(keras.layers.Dense(units = 48, activation='relu'))
# 48 * (64+ 1)
animal_trad.add(keras.layers.Dense(units = 32, activation='relu'))
# 32*(48 + 1)
animal_trad.add(keras.layers.Dense(units = 16, activation='relu'))
# 16*(32 + 1)
animal_trad.add(keras.layers.Dense(units=3, activation='softmax'))
# 3*(16+1)
animal_trad.summary()
```

```python
animal_trad.compile(
    loss=keras.losses.CategoricalCrossentropy(),
    optimizer=keras.optimizers.RMSprop(learning_rate=0.001),
    metrics= keras.metrics.CategoricalAccuracy()
)
```

```python
result1 = animal_trad.fit_generator(
    generator= train_generator,
    epochs=15,
    steps_per_epoch=14630//64,
    validation_steps=1500/64,
    validation_data=test_generator)
```

```python
result1.history.keys()
```

```python
plt.plot(result1.history['categorical_accuracy'], 'go', label='Training accuracy')
plt.plot(result1.history['val_categorical_accuracy'], 'g-', label='Validation accuracy')
plt.title('Validation curve')
plt.xlabel('Epoches')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(result1.history['loss'], 'go', label='Training loss')
plt.plot(result1.history['val_loss'], 'g-', label='Validation loss')
plt.title('Validation curve')
plt.xlabel('Epoches')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
animal_conv = keras.models.Sequential()
animal_conv.add(keras.layers.InputLayer(input_shape=(126, 126, 3)))
animal_conv.add(keras.layers.Conv2D(filters = 32, kernel_size = (3,3), activation='relu'))
#32*(3*3*3 + 1)
animal_conv.add(keras.layers.MaxPooling2D(pool_size = (2,2), padding='same', strides=(2, 2)))
animal_conv.add(keras.layers.Conv2D(filters = 48, kernel_size = (3,3), activation='relu',
                                kernel_regularizer = keras.regularizers.l2(0.001)))
# 48*(32*3*3+1)
animal_conv.add(keras.layers.MaxPooling2D(pool_size = (2,2), padding='same', strides=(2, 2))) 
animal_conv.add(keras.layers.Conv2D(filters = 64, kernel_size = (3,3), activation='relu',
                                kernel_regularizer = keras.regularizers.l2(0.001)))
# 64*(48*3*3+1)
animal_conv.add(keras.layers.MaxPooling2D(pool_size = (2,2), padding='same', strides=(2, 2))) 
animal_conv.add(keras.layers.Conv2D(filters = 96, kernel_size = (3,3), activation='relu',
                                kernel_regularizer = keras.regularizers.l2 (0.001)))
# 96*(64*3*3+1)
animal_conv.add(keras.layers.MaxPooling2D(pool_size = (2,2), padding='same', strides=(2, 2))) 
animal_conv.add(keras.layers.Flatten())   # 6*6* 96
animal_conv.add(keras.layers.Dense(units=64, activation='relu'))
#(6*6* 96 +1) *(64)
animal_conv.add(keras.layers.Dense(units=3, activation='softmax'))
3*(64 + 1)
animal_conv.summary()
```
